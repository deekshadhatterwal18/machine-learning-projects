```python
import face_recognition
import numpy as np
from pathlib import Path
import pickle


data = pickle.load(open("encoded_faces/encodings.pkl", "rb"))
known_encodings = data["encodings"]
known_names = data["names"]

TEST_DIR = Path("test_images")

correct = 0
total = 0

for s in TEST_DIR.iterdir():
    if not s.is_dir():
        continue

    actual = s.name

    for img_path in s.iterdir():
        img = face_recognition.load_image_file(img_path)
        enc = face_recognition.face_encodings(img)

        if not enc:
            print("No face in", img_path.name)
            continue

        total += 1
        enc = enc[0]

        distances = face_recognition.face_distance(known_encodings, enc)
        idx = np.argmin(distances)
        predicted = known_names[idx]

        print(img_path.name, "→ Pred:", predicted, "| Actual:", actual)

        if predicted == actual:
            correct += 1

accuracy = (correct / total * 100) if total > 0 else 0
print("\nAccuracy =", accuracy, "%")
```
